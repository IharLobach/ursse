```python
import ursse.sync_motion_simulation as sms
import pandas as pd
import os 
import numpy as np
from ursse.path_assistant import PathAssistant
```

```python
rf_noise_std_list = 1e-5*np.arange(16)
str_label_list = [f'{x:.1e}' for x in rf_noise_std_list]
```

```python
for i, rf_noise_std, str_label in zip(range(len(rf_noise_std_list)),
                                      rf_noise_std_list, str_label_list):
    print(f'-----------Working on {i+1} out of {len(rf_noise_std_list)}-----------')
    model = sms.Model(rf_noise_std=rf_noise_std)
    res = model.simulate()
    for f in model.files_and_pars:
        f['sim_df_before_spad'].to_pickle(
            os.path.join(PathAssistant(f['shift']).get_shift_cache_folder_path(),
            f['file'].replace(".ptu", f"_{str_label}_sim.pkl")))
```

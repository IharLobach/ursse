```python
import ursse.sync_motion_simulation as sms
import pandas as pd
import os 
import numpy as np
from ursse.path_assistant import PathAssistant
import matplotlib.pyplot as plt
from datetime import datetime
import pickle
```

```python
rf_noise_std_list = 1e-5*np.arange(16)
str_label_list = [f'{x:.1e}' for x in rf_noise_std_list]
spad_tts_list = [0.350, 0.400, 0.450]
spad_tts_dist_list = ['Gamma', 'Gaussian']
tot_n = len(str_label_list) * len(spad_tts_list) * len(spad_tts_dist_list)
all_results = []
i = 0
for rf_noise_std, str_label in zip(rf_noise_std_list, str_label_list):
    model = sms.Model(rf_noise_std=rf_noise_std)
    model.simulate(load=str_label)
    for dist in spad_tts_dist_list:
        model.spad_tts_distribution = "Gamma"
        for tts in spad_tts_list:
            i += 1
            print(f"---------Working on {i} out of {tot_n}, {datetime.now()}---------")
            model.spad_tts = tts
            model.add_spad_tts_do_fitting_and_binning(verbose=True)
            res = {}
            features = ['amplitude', 'rms_length', 'slow_phase', 'kick_to_amplitude']
            nbins_arr = [40, 40, 20, 40]
            for feat, nbins in zip(features, nbins_arr):
                res[feat] = model.get_meas_sim_comparison(feature=feat, nbins=nbins)
            res['amp_per_corr'] = model.get_meas_sim_amp_period_corr()
            all_results.append({"rf_noise_std": rf_noise_std,
                                "rf_std_label": str_label,
                                "spad_tts": tts,
                                "spad_tts_dist": dist,
                                "data": res})
            with open("all_results.pkl", 'wb') as f:
                pickle.dump(all_results, f)
```

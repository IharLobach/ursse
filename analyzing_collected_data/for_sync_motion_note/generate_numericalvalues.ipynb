```python
from config_ursse import get_from_config as get
import os
```

```python
def newval(name, val, unit):
    return (r'\newcommand{\val'+name+r'}{\SI{'
              +val
              +r'}{'
              +unit
              +r'}}'+'\n')
```

```python
gamma =  get("gamma")
me = get("me_MeV")
ring_alpha = get("ring_alpha")
```

```python
res = \
    r'% Do not edit this file. It is generated by a Python script.'+'\n'\
    +r'\ProvidesPackage{numericalvalues}'+'\n'\
    +r'\usepackage{siunitx}'+'\n'+'\n'+'\n'

res += newval("nav", f'{get("nav"):.1f}', "")
res += newval("Vrf", f'{get("Vrf"):.0f}', "V")
res += newval("gamma", f'{gamma:.1f}', "")
res += newval("Ezero", f'{gamma*me:.1f}', "MeV")
res += newval("JE", f'{get("damping_partition_JE"):.2f}', "")
res += newval("alphac", f'{ring_alpha:.5f}', "")
res += newval("etas", f'{ring_alpha-1/gamma**2:.5f}', "")
res += newval("phis", f'{get("phis"):.4f}', "rad")
res += newval("Uav", f'{get("Et"):.1f}', "eV")
res += newval("h", f'{get("RF_q"):.0f}', "")
res += newval("Ec", f'{get("Ec"):.1f}', "eV")
res += newval("rho", f'{100*get("dipole_rho_m"):.0f}', "cm")
res += newval("M", f'{get("M"):.2f}', "")

print(res)
```

```python
with open(
    os.path.join(
        get("URSSE_sync_motion_note_dir"), "numericalvalues.sty"), 'w') as f:
    f.write(res)
```
